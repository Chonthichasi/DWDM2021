# นำเข้าข้อมูล alldb.csv 

```python
import pandas as pd # ทำงานกับข้อมูลที่เป็นรูปแบบตาราง
import os # จัดการไฟล์และpath ที่นำเข้า
```

```python
from google.colab import drive # เชื่อมgoogle drive
drive.mount('/content/drive') 
path='/content/drive/My Drive/Final data 2022'
```

```python
data = pd.read_csv(os.path.join(path,'alldb.csv')) #อ่านไฟล์ csv เเล้วชี้ไปที่ไฟล์ alldb.csv
data # เรียกดูตาราง
```

## ตรวจสอบว่ามีค่า Missing หรือไม่

```python
data.isnull().any() #ดูค่า Missing รายคอลัมน์ของตาราง data
#เนื่องจากมี True เกือบทุกคอลัมน์ แสดงว่ามีค่าที่เป็น Missing 
```

## เลือกเฉพาะคอลัมน์ที่ต้องการ

```python
data_all = data[['user','RowKey','date','phone','site','time','urine','ov','PartitionKey','title','age']] # เลือกเฉพาะคอลัมน์ที่ต้องการ คือ user,RowKey,date,phone,site,time,urine,ov,PartitionKey,title เเละ age
data_all # เรียกดูตารางข้อมูล data_all
```

## user เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
data_alldb = data_all[(data_all['user'].astype(str).str.isdigit())& (data_all['user'].str.len() == 5)]  # สร้างตัวแปร data_alldb มารับคอลัมน์ user โดยใช้คำสั่ง str.isdigit() เพื่อตรวจสอบว่าเป็นตัวเลขหรือไม่ เเละเปลียนคอลัมน์ user เป็นตัวเลข 5 หลัก
data_alldb
```

```python
data_alldb.isnull().any() #ตรวจสอบ Missing คอลัมน์ user ของตาราง data_alldb
#คอลัมน์ user เป็นค่า False แสดงว่า ไม่มีค่า Missing
```

```python
set(data_alldb['user']) # เรียกดู set ข้อมูลของ user 
```

## RowKey เป็น String

```python
data_alldb['RowKey'] = data_alldb['RowKey'].astype('str') # ใช้คำสั่ง .astype เปลียนชนิดข้อมูลในคอลัมน์ RowKey เป็น String
data_alldb
```

```python
data_alldb.isnull().any() #ตรวจสอบ Missing คอลัมน์ RowKey ของตาราง data_alldb
#คอลัมน์ RowKey เป็นค่า False แสดงว่า ไม่มีค่า Missing
```

## date เป็น YYYY-MM-DD

```python
data_alldb['date']=pd.to_datetime(data_alldb['date'], format='%Y-%m-%d') # จัดรูปแบบการแสดงผลของคอลัมน์ date เป็น YYYY-MM-DD
data_alldb
```

```python
data_alldb.isnull().any() #ตรวจสอบ Missing คอลัมน์ date ของตาราง data_alldb
# พบว่าคอลัมน์ date เป็น False เเสดงว่าไม่มีค่า Missing 
```

```python
data_alldb['date'] = data_alldb[['date']].fillna(method="ffill")  # เติมค่าว่างของวันที่ด้วย ค่าก่อนหน้า หรือค่าข้างบน    
data_alldb
```

```python
data_alldb.isnull().any() #ตรวจสอบ Missing คอลัมน์ของตาราง data_alldb
```

## phone เป็น String

```python
data_alldb['phone'] = data_alldb['phone'].astype('str') # ใช้คำสั่ง .astype เปลียนชนิดข้อมูลในคอลัมน์ phone เป็น String
data_alldb
```

## site เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
data_alldb = data_alldb[(data_all['site'].astype(str).str.isdigit())& (data_all['site'].str.len() == 5)] # สร้างตัวแปร data_alldb มารับคอลัมน์ site โดยใช้คำสั่ง str.isdigit() เพื่อตรวจสอบว่าเป็นตัวเลขหรือไม่ เเละเปลียนคอลัมน์ site เป็นตัวเลข 5 หลัก
data_alldb
```

```python
set(data_alldb['site']) # เรียกดู set ข้อมูลของ site
```

## time เป็น hh-mm-ss

```python
data_alldb['time']=data_alldb['time'].replace(':','-',regex=True) # ใช้คำสั่ง .replace จัดรูปแบบการแสดงผลของคอลัมน์ time เป็น hh-mm-ss
data_alldb
```

```python
set(data_alldb['time']) # เรียกดู set ข้อมูลของ time 
```

## urine มีค่า String

```python
data_alldb['urine'] = data_alldb['urine'].astype('str') # ใช้คำสั่ง .astype เปลียนชนิดข้อมูลในคอลัมน์ urine เป็น String
data_alldb
```

```python
set(data_alldb['urine']) # เรียกดู set ข้อมูลของ urine
```

## ov เป็นตัวเลข มีค่าตั้งแต่ 0-4 
(0 คือไม่พบโรค 1-4 คือพบโรคที่ระดับความรุนแรงต่างๆ)

```python
import numpy as np # นำเข้า numpy ในชื่อ np
```

```python
set(data_alldb['ov']) # เรียกดู set ข้อมูลของ ov
```

```python
data_alldb['ov'] = pd.to_numeric(data_alldb['ov'], errors='coerce') # เปลี่ยนคอลัมน์ ov ให้เป็นตัวเลข
data_alldb = data_alldb.replace(np.nan, 0, regex=True)    # เปลี่ยนค่าว่างให้เป็น 0
data_alldb['ov'] = data_alldb['ov'].astype(int)    # เปลี่ยน type ของ ov ให้เป็น int
```

```python
set(data_alldb['ov']) # เรียกดู set ข้อมูลของ ov อีกครั้ง
```

```python
data_alldb = data_alldb[(data_all['PartitionKey'].astype(str).str.isdigit())& (data_all['PartitionKey'].str.len() == 5)] # สร้างตัวแปร data_alldb มารับคอลัมน์ PartitionKey โดยใช้คำสั่ง str.isdigit() เพื่อตรวจสอบว่าเป็นตัวเลขหรือไม่ เเละเปลียนคอลัมน์ user เป็นตัวเลข 5 หลัก
data_alldb
```
