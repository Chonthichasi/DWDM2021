````python
set (airports['รัฐ']) # ดูว่ามีรัฐอะไรบ้าง
````

````python
len(set(airports['รัฐ'])) # ดูว่ามีทั้งหมดกี่รัฐ
````

````python
flights_clean = flights[flights['ความล่าช้าขาออก'] >= 0] # เลือกเอาข้อมูลที่ ความล่าช้าขาออก มากกว่าหรือเท่ากับ 1
flights_clean
````

````python
flights_clean = flights[flights['ความล่าช้าขาเข้า'] >= 0] # เลือกเอาข้อมูลที่ ความล่าช้าขาเข้า มากกว่าหรือเท่ากับ 1
flights_clean
````

````python
merged_table = flights.merge(airports,how='left',left_on='รหัสสนามบินต้นทาง',right_on='รหัสสนามบิน')
merged_table
````

````python
merged_table['ความล่าช้าขาออก'].max()
````

````python
max = merged_table.groupby('รัฐ').max().sort_values(by=['ความล่าช้าขาออก'],ascending=False)
max.iloc[:1,:]
````

````python
merged_table['ความล่าช้าขาออก'].min()
````

````python
min = merged_table.groupby('รัฐ').min().sort_values(by=['ความล่าช้าขาออก'],ascending=False)
min.iloc[:1,:]
````
